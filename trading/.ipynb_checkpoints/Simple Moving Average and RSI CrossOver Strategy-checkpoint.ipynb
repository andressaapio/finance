```python
import yfinance as yf
import matplotlib.pyplot as plt
import datetime
import pandas as pd
import numpy as np
```

## Configurações iniciais

```python
start_date = datetime.datetime(2010, 1, 1)
end_date = datetime.datetime(2022, 1, 1)
stock = 'IBM' #  [' MSFT, AAPL, VALE3.SA', PETR4.SA, MGLU3.SA
short_period = 40
long_period = 150
capital = 100
is_long = False
```

## Download dos dados

```python
stock_data = {}
ticker = yf.download(stock, start_date, end_date)
stock_data['price'] = ticker['Adj Close']
data = pd.DataFrame(stock_data)
```

## Construção das médias móveis

```python
data['short_ma'] = data['price'].ewm(span=short_period).mean()
data['long_ma'] = data['price'].ewm(span=long_period).mean()

## RSI
data['move'] = data['price'] - data['price'].shift(1)
data['up'] = np.where(data['move'] > 0, data['move'], 0)
data['down'] = np.where(data['move'] < 0, data['move'], 0)

data['average_gain'] = data['up'].rolling(14).mean()
data['average_lost'] = data['down'].abs().rolling(14).mean()

relative_strength = data['average_gain'] / data['average_lost']
data['rsi'] = 100.0 - (100.0 / (1.0 + relative_strength))

data = data.dropna()
```

```python
data
```

## Estratégia de compra e venda de ativos

```python
def strategy(data, capital, is_long):
    # we consider all the trading days and decide whether to open a 
    # long position or not
    
    equity = [capital]
    price_when_buy = 0

    for index, row in data.iterrows():
        # CLOSE THE LONG POSITION WE HAVE OPENED
        if row['short_ma'] < row['long_ma'] and is_long: # is_long = True
            # row.price / price_when_buy -> se > 1, ganhou dinheiro, se < 1, perdeu dinheiro
            equity.append(capital * row.price / price_when_buy) 
            is_long = False

        elif row['short_ma'] > row['long_ma'] and not is_long and row['rsi'] < 30: # is_long = False
            # OPEN A LONG POSITION
            price_when_buy = row['price']
            is_long = True
    
    return equity
```

```python
equity = strategy(data, capital, is_long)
```

## Plotar gráficos

```python
plt.figure(figsize=(12,6))
plt.plot(data['price'], label='Stock Price')
plt.plot(data['short_ma'], label='Short MA', color='blue')
plt.plot(data['long_ma'], label='Long MA', color='green')
plt.title('Moving Average (MA) Crossover Trading Strategy')
plt.xlabel('Date')
plt.ylabel('Stock price')
plt.show()
```

```python
print("Profit of the trading strategy: %.2f%%" % (
    (float(equity[-1]) - float(equity[0])) /
    float(equity[0]) * 100))
print("Actual capital: $%0.2f" % equity[-1])


plt.figure(figsize=(12,6))
plt.title('Equity Curve')
plt.plot(equity, label='Stock Price', color='green')
plt.xlabel('Date')
plt.ylabel('Actual Capital ($)')
plt.show()
```
