```python
import datetime
import pandas as pd
import yfinance as y
import matplotlib.pyplot as plt
import numpy as np
```

```python
start = datetime.datetime(2020, 1, 1)
end = datetime.datetime(2021, 1, 1)
```

```python
papeis = ['petr4.sa', 'vale3.sa', 'itub4.sa', 'bpac11.sa']
```

```python
ibov = y.download('^bvsp', start, end)['Adj Close']
```

```python
ibov
```

```python
carteira = y.download(papeis, start, end)['Adj Close']
```

```python
carteira
```

```python
(carteira / carteira.iloc[0]).plot(figsize=(13,8))
(ibov / ibov.iloc[0]).plot(label = 'IBOV').legend()
```

```python
rentabilidade = carteira.pct_change() * 100
rentabilidade_anual = np.sum(rentabilidade) # rentabilidade.mean() * 250
```

```python
rentabilidade_anual
```

```python
rentabilidade_ibov = ibov.pct_change() * 100
rentabilidade_ibov_anual = np.sum(rentabilidade_ibov) # rentabilidade.mean() * 250
```

```python
rentabilidade_ibov_anual
```

```python
peso = np.array([0.25, 0.25, 0.25, 0.25])
```

```python
rentabilidade_carteira = np.dot(rentabilidade_anual, peso)
```

```python
rentabilidade_carteira
```
